```python
#hinh 3
import turtle
a = turtle.Turtle()
c = 200

a.pencolor('blue')
a.forward(c)
a.left(90)
a.forward(c)
a.left(90)
a.forward(c)
a.left(90)
a.forward(c)
a.left(90)

a.penup()
a.goto(25,25)
a.pendown()
a.pencolor('yellow')
a.forward(c-50)
a.left(90)
a.forward(c-50)
a.left(90)
a.forward(c-50)
a.left(90)
a.forward(c-50)
a.left(90)

a.penup()
a.goto(50,50)
a.pendown()
a.pencolor('red')
a.forward(c-100)
a.left(90)
a.forward(c-100)
a.left(90)
a.forward(c-100)
a.left(90)
a.forward(c-100)
a.left(90)

turtle.done()
```
