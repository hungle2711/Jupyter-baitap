```python
item_list = {}

def add(item_list):
    print('*** Chức năng thêm sản phẩm ***')
    id_item = input('Nhập mã sản phẩm: ')
    name = input('Nhập tên sản phẩm: ')
    if id_item in item_list:
        print(f'Sản phẩm với mã {id_item} đã tồn tại.')
    else:
        item_list[id_item] = {
            'ten' : name
        }
        print(f'thêm sản phẩm thành công với mã {id_item}.')

def fix(item_list):
    print('***Chức năng sửa sản phẩm***')
    id_item = input('Nhập mã sản phẩm cần sửa: ')
    if id_item not in item_list:
        print('Sản phẩm không tồn tại')
        return
    name = input('Nhập tên mới cho sản phẩm: ')
    item_list[id_item] = {
        'ten' :name
    }
    print(f'Cập nhật thành công sản phẩm có mã là {id_item}')

def delete(item_list):
    print('***Chức năng xóa sản phẩm***')
    id_item = input('Nhập mã sản phẩm cần xóa: ')
    if id_item not in item_list:
        print('Sản phẩm không tồn tại')
        return
    del item_list[id_item]
    print(f'Xóa thành công sản phẩm với mã {id_item}')

def show(item_list):
    print('***Chức năng hiển thị danh sách sản phẩm***')
    for id_item, info in item_list.items():
        print(f'Mã: {id_item}.  Tên: {info["ten"]}')

while True:
    print('=== Chương trình quản lý sản phẩm ===')
    print('Chọn thao tác mà bạn muốn: ')
    print('1. Thêm sản phẩm')
    print('2. Sửa sản phẩm')
    print('3. Xóa sản phẩm')
    print('4. Hiển thị danh sách sản phẩm')
    print('5. Thoát')
    chose = input('Nhập thao tác mà bạn muốn(1-5): ')
    if chose == '1':
        add(item_list)
    elif chose == '2':
        fix(item_list)
    elif chose == '3':
        delete(item_list)
    elif chose == '4':
        show(item_list)
    elif chose == '5':
        print('Hẹn gặp lại lần sau')
        break
    else:
        print('Thao tác không tồn tại')
```
