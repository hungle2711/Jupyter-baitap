```python
#hinh1
import turtle

canh = 200
a=turtle.Turtle()
a.pensize(3)
a.penup()
a.goto(-100,-100)
a.pendown()
a.forward(canh)
a.left(90)
a.forward(canh)
a.left(90)
a.forward(canh)
a.left(90)
a.forward(canh)
a.left(90)
a.penup()
a.goto(0,142)
a.right(45)
a.pendown()
a.forward(canh)
a.right(90)
a.forward(canh)
a.right(90)
a.forward(canh)
a.right(90)
a.forward(canh)
a.right(90)
a.hideturtle()
turtle.done()
```

```python
#hinh2
import turtle
canh = 200
nua_canh = 100
a=turtle.Turtle()
a.pensize(3)
a.forward(canh)
a.left(90)
a.forward(canh)
a.left(90)
a.forward(canh)
a.left(90)
a.forward(canh)
a.left(90)
a.penup()
a.goto(nua_canh,canh)
a.pendown()
a.right(90)
a.forward(nua_canh)
a.left(90)
a.forward(canh)
a.left(90)
a.forward(canh)
a.left(90)
a.forward(canh)
a.left(90)
a.forward(nua_canh)
a.hideturtle()
turtle.done()
```
