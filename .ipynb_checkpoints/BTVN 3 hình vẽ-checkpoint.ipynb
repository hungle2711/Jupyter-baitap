```python
#hinh1
import turtle
a=turtle.Turtle()
a.pensize(3)
a.penup()
a.goto(-100,-100)
a.pendown()
a.forward(200)
a.left(90)
a.forward(200)
a.left(90)
a.forward(200)
a.left(90)
a.forward(200)
a.left(90)
a.penup()
a.goto(0,142)
a.right(45)
a.pendown()
a.forward(200)
a.right(90)
a.forward(200)
a.right(90)
a.forward(200)
a.right(90)
a.forward(200)
a.right(90)
a.hideturtle()
turtle.mainloop()
```
