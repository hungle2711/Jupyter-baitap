```python
# Create a scene with a house, a tree, and a sun
# Based on the picture at http://dragonometry.net/blog/?p=566

import turtle
import math

# Set the background color
screen = turtle.Screen()
screen.bgcolor("skyblue")

# Create our turtle
house = turtle.Turtle()
house.color("black")
house.shape("turtle")
house.speed(10)

# Define a funtion to draw and fill a rectangle with the given
# dimensions and color
def ve_hcn(t, width, height, color):
  t.fillcolor(color)
  t.begin_fill()
  t.forward(width)
  t.left(90)
  t.forward(height)
  t.left(90)
  t.forward(width)
  t.left(90)
  t.forward(height)
  t.left(90)
  t.end_fill()
  
# Define a function to draw and fill an equalateral right 
# triangle with the given hypotenuse length and color.  This
# is used to create a roof shape.
def ve_htg(t, length, color):
  t.fillcolor(color)
  t.begin_fill()
  t.forward(length)
  t.left(135)
  t.forward(length / math.sqrt(2))
  t.left(90)
  t.forward(length / math.sqrt(2))
  t.left(135)
  t.end_fill()
  
# Define a function to draw and fill a parallelogram, used to
# draw the side of the house
def ve_hbh(t, width, height, color):
  t.fillcolor(color)
  t.begin_fill()
  t.left(30)
  t.forward(width)
  t.left(60)
  t.forward(height)
  t.left(120)
  t.forward(width)
  t.left(60)
  t.forward(height)
  t.left(90)
  t.end_fill()
  
# Define a function to draw four sun rays of the given length,
# for the sun of the given radius.  The turtle starts in the 
# center of the circle.
def ve_4tia(t, length, radius):
  for i in range(4):
    t.penup()
    t.forward(radius)
    t.pendown()
    t.forward(length)
    t.penup()
    t.backward(length + radius)
    t.left(90)

# Draw and fill the front of the house
house.penup() 
house.goto(-150, -120)
house.pendown()
ve_hcn(house, 100, 110, "blue")

# Draw and fill the front door
house.penup()
house.goto(-120, -120)
house.pendown()
ve_hcn(house, 40, 60, "lightgreen")

# Front roof
house.penup()
house.goto(-150, -10)
house.pendown()
ve_htg(house, 100, "magenta")

# Side of the house
house.penup()
house.goto(-50, -120)
house.pendown()
ve_hbh(house, 60, 110, "yellow")

# Window
house.penup()
house.goto(-30, -60)
house.pendown()
ve_hbh(house, 20, 30, "brown")

# Side roof
house.penup()
house.goto(-50, -10)
house.pendown()
house.fillcolor("orange")
house.begin_fill()
house.left(30)
house.forward(60)
house.left(105)
house.forward(100 / math.sqrt(2))
house.left(75)
house.forward(60)
house.left(105)
house.forward(100 / math.sqrt(2))
house.left(45)
house.end_fill()

# Tree base
house.penup() 
house.goto(100, -130)
house.pendown()
ve_hcn(house, 20, 40, "brown")

# Tree top
house.penup() 
house.goto(65, -90)
house.pendown()
ve_htg(house, 90, "lightgreen")
house.penup() 
house.goto(70, -45)
house.pendown()
ve_htg(house, 80, "lightgreen")
house.penup() 
house.goto(75, -5)
house.pendown()
ve_htg(house, 70, "lightgreen")

# Sun
house.penup() 
house.goto(55, 110)
house.fillcolor("yellow")
house.pendown()
house.begin_fill()
house.circle(24)
house.end_fill()

# Sun rays
house.penup()
house.goto(55, 134)
house.pendown()
ve_4tia(house, 25, 24)
house.right(45)
ve_4tia(house, 15, 24)
house.left(45)

# Put down some labels
house.color("black")
house.penup()
house.goto(-150, 70)
house.pendown()
house.write("House", "14pt bold")
house.penup()
house.goto(150, -150)
house.pendown()
house.write("Tree", "14pt bold")
house.penup()
house.goto(130, 150)
house.pendown()
house.write("Sun", "14pt bold")

# Bring the turtle down to the front door, and we're done!
house.penup()
house.goto(-100, -150)
house.left(90)
turtle.done()
```
