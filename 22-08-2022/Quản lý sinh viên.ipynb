```python
import tkinter as tk
from tkinter import ttk
from openpyx1 import load_workbook




main = tk.Tk()
main.title("Ứng dụng quản lý sinh viên")



















class SinhVien:
    
    def __init__(self, ma, ten, tuoi, lop):
        self.ma = ma
        self.ten = ten
        self.tuoi = tuoi
        self.lop = lop
        
class DanhSachSV:
     
    def __init__(self):
        self.danh_sach = []

    def them(self):
        print('***Chức năng thêm sinh viên***')
        ma = input('Nhập mã sinh viên: ')
        ten = input('Nhập tên sinh viên: ')
        tuoi = input('Nhập tuổi sinh viên: ')
        lop = input('Nhập lớp sinh viên: ')  
        if ma in self.danh_sach:
            print('Sinh viên đã tồn tại')
        else:
            sinh_vien = SinhVien(ma, ten, tuoi, lop)
            self.danh_sach.append(sinh_vien)
            print(f'Thêm thành công sinh viên có mã: {ma}')
        
    def hien_thi(self):
        print('***Chức năng hiển thị danh sách sinh viên***')
        print('Mã\t\t Tên\t\t Tuổi\t\t Lớp')
        for sinh_vien in self.danh_sach:
            print(f'{sinh_vien.ma}\t\t {sinh_vien.ten}\t\t {sinh_vien.tuoi}\t\t {sinh_vien.lop}\t\t')
    
    
    def sua(self):
        print('***Chức năng sửa thông tin sinh viên***')
        ma = input('Nhập mã sinh viên bạn muốn sửa')
        sinh_vien = None
        for sv in self.danh_sach:
            if sv.ma == ma:
                sinh_vien = sv
                break
        if sv == None:
            print(f'Sinh viên với mã {sv.ma} không tồn tại')
            return
        else:
            ten = input('Nhập tên: ')
            tuoi = input('Nhập tuổi: ')
            lop = input('Nhập lớp: ')
            sinh_vien.ten = ten
            sinh_vien.tuoi = tuoi
            sinh_vien.lop = lop
        print('Cập nhật thành công!')
    
    
    
    def tim_kiem(self):
        print('***Chức năng tìm kiếm sinh viên***')
        ma = input('Nhập mã sinh viên mà bạn muốn tìm kiếm: ')
        for sv in self.danh_sach:
            if sv.ma == ma:
                print('Mã\t\t Tên\t\t Tuổi\t\t Lớp')
                print(f'{sv.ma}\t\t {sv.ten}\t\t {sv.tuoi}\t\t {sv.lop}\t\t')
        if ma not in self.danh_sach:
            print('Sinh viên không tồn tại!')
        
        
    def xoa(self):
        print('***Chức năng xóa sinh viên***')
        ma = input('Nhập mã sinh viên mà bạn muốn xóa: ')
        for sv in self.danh_sach:
            if sv.ma == ma:
                self.danh_sach.remove(sv)
                print(f'Đã xóa thành công sinh viên!')
        if ma not in self.danh_sach:
            print('Sinh viên không tồn tại')
            
            
    def sap_xep(self):
        print('***Chức năng sắp xếp danh sách***')
        while True:
            print('Lựa chọn sắp xếp: \n\
            1.Sắp xếp theo mã! \n\
            2.Sắp xếp theo tên! \n\
            3.Sắp xếp theo tuổi! \n\
            4.Sắp xếp theo lớp! \n\
            5.Thoát!')
            lua_chon = input('Lựa chọn thao tác mà bạn muốn dùng ở trên (1-5): ')
            if lua_chon == '1':
                self.danh_sach.sort(key=lambda sinh_vien:sinh_vien.ma)
            elif lua_chon == '2':
                self.danh_sach.sort(key=lambda sinh_vien:sinh_vien.ten)
            elif lua_chon == '3':
                self.danh_sach.sort(key=lambda sinh_vien:sinh_vien.tuoi)
            elif lua_chon == '4':
                self.danh_sach.sort(key=lambda sinh_vien:sinh_vien.lop)
            elif lua_chon == '5':
                return
# Sau khi trọn 1-4 thì trọn 5 rồi ra chọn hiển thị danh sách!!!
    
class MainApp: 
    def run(self):
        danh_sach_sv = DanhSachSV()
        
        while True:
            print('===Chương trình quản lý sinh viên===')
            print('1. Xem danh mục sinh viên')
            print('2. Thêm sinh viên mới vào danh mục')
            print('3. Cập nhật thông tin của sinh viên trong danh mục')
            print('4. Xóa thông tin sinh viên khỏi danh mục')
            print('5. Tìm kiếm thông tin sinh viên trong danh mục theo từ khóa')
            print('6. Sắp xếp thông tin sinh viên trong danh mục')
            print('7. Thoát')
            lua_chon = input('Chọn chức năng trên: ')
            if lua_chon == '1':
                danh_sach_sv.hien_thi()
            elif lua_chon == '2':
                danh_sach_sv.them()
            elif lua_chon == '3':
                danh_sach_sv.sua()
            elif lua_chon == '4':
                danh_sach_sv.xoa()
            elif lua_chon == '5':
                danh_sach_sv.tim_kiem()
            elif lua_chon == '6':
                danh_sach_sv.sap_xep()
                danh_sach_sv.hien_thi()
            elif lua_chon == '7':
                return
            
#Giao diện
#row 1










            
            
chuong_trinh_chinh = MainApp()
chuong_trinh_chinh.run()
```
