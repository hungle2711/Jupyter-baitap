```python
import tkinter as tk
from tkinter import ttk


expression = ''


def update_expression_input(exp):
    expression_input.delete(0, tk.END)
    expression_input.insert(0, exp)


def handle_click(value):
    global expression
    expression += value
    update_expression_input(expression)
    
    
def handle_click_equal():
    global expression
    try:
        update_expression_input(eval(expression))
    except:
        update_expression_input('Error')
    expression = ''
 
 
def handle_click_clear():
    global expression
    update_expression_input(' ')
    expression = ''

main = tk.Tk()
main.title("Calculate")
    
    
# Row1
expression_input = ttk.Entry(main)
expression_input.grid(row=0,column=0, columnspan =4, sticky='we')

# Row2
btn1 = ttk.Button(main, text='1', command= lambda:handle_click('1'))
btn2 = ttk.Button(main, text='2', command= lambda:handle_click('2'))
btn3 = ttk.Button(main, text='3', command= lambda:handle_click('3'))
btn_pls = ttk.Button(main, text='+', command= lambda:handle_click('+'))

btn1.grid(row=1, column=0)
btn2.grid(row=1, column=1)
btn3.grid(row=1, column=2)
btn_pls.grid(row=1, column=3)

# Row3
btn4 = ttk.Button(main, text='4', command= lambda:handle_click('4'))
btn5 = ttk.Button(main, text='5', command= lambda:handle_click('5'))
btn6 = ttk.Button(main, text='6', command= lambda:handle_click('6'))
btn_mn = ttk.Button(main, text='-', command= lambda:handle_click('-'))

btn4.grid(row=2, column=0)
btn5.grid(row=2, column=1)
btn6.grid(row=2, column=2)
btn_mn.grid(row=2, column=3)

# Row4
btn7 = ttk.Button(main, text='7', command= lambda:handle_click('7'))
btn8 = ttk.Button(main, text='8', command= lambda:handle_click('8'))
btn9 = ttk.Button(main, text='9', command= lambda:handle_click('9'))
btn_time = ttk.Button(main, text='*', command= lambda:handle_click('*'))

btn7.grid(row=3, column=0)
btn8.grid(row=3, column=1)
btn9.grid(row=3, column=2)
btn_time.grid(row=3, column=3)

# Row5
btn0 = ttk.Button(main, text='0', command= lambda:handle_click('0'))
btn_clear = ttk.Button(main, text='Clear', command=handle_click_clear)
btn_equal = ttk.Button(main, text='=', command=handle_click_equal)
btn_div = ttk.Button(main, text='/', command= lambda:handle_click('/'))

btn0.grid(row=4, column=0)
btn_clear.grid(row=4, column=1)
btn_equal.grid(row=4, column=2)
btn_div.grid(row=4, column=3)

# Row6
btn_dot = ttk.Button(main, text='.', command= lambda:handle_click('.'))

btn_dot.grid(row=5, column=0)

main.mainloop()
```
