```python
import turtle
import math

x = int(math.sqrt((200**2)/2))
screen=turtle.Screen()
screen.setup(900,600)
pen=turtle.Turtle()
pen.pensize(2)
pen.speed(5)
pen.color('gray')

pen.forward(200)
pen.right(90)
pen.forward(100)
pen.right(90)
pen.forward(200)
pen.right(90)
pen.forward(100)
pen.right(90)
pen.left(45)
pen.forward(x)
pen.right(90)
pen.forward(x)
pen.penup()
pen.backward(x/2)
pen.pendown()
pen.left(135)
pen.forward(50)
pen.right(90)
pen.forward(25)
pen.right(90)
pen.forward(75)
pen.penup()
pen.goto(0,0)
pen.pendown()
pen.left(90)
pen.forward(50)
pen.right(90)
pen.forward(100)
pen.left(90)
pen.forward(50)
pen.left(90)
pen.forward(100)

class caythong():
    """Xay dung ti le cho cay thong"""	
    def __init__(self,x=0,y=0,h=0):	
        self.x=x
        self.y=y
        self.h=h
        k=h/250
        self.k=k
        # Create all point off tree : A1 A2 ... A13	
        self.A1=(self.x,self.y)	
        self.A2=(int(x+40*k),y)	
        self.A3=(int(x-60*k),int(y+80*k))
        self.A4=(int(x+ 8*k),int(y+80*k))
        self.A5=(int(x+32*k),int(y+80*k))
        self.A6=(int(x+100*k),int(y+80*k))
        self.A7=(int(x-40*k),int(y+150*k))
        self.A8=(int(x+20*k),int(y+150*k))
        self.A9=(int(x+80*k),int(y+150*k))
        self.A10=(int(x-20*k),int(y+200*k))
        self.A11=(int(x+20*k),int(y+200*k))
        self.A12=(int(x+60*k),int(y+200*k))
        self.A13=(int(x+20*k),int(y+250*k))

    def paint(self):
        # Ve than cay
        pen.penup()
        pen.goto(self.A1)
        pen.pendown()
        pen.fillcolor('brown')
        pen.begin_fill()
        pen.goto(self.A2)
        pen.goto(self.A5)
        pen.goto(self.A4)
        pen.goto(self.A1)
        pen.end_fill()
        # Ve ngon cay 1/3
        pen.penup()
        pen.goto(self.A3)
        pen.pendown()
        pen.fillcolor('#525f30')
        pen.begin_fill()
        pen.goto(self.A6)
        pen.goto(self.A8)
        pen.goto(self.A3)
        pen.end_fill()
        # Ve ngon cay 2/3
        pen.penup()
        pen.goto(self.A7)
        pen.pendown()
        pen.fillcolor('#7c9048')
        pen.begin_fill()
        pen.goto(self.A9)
        pen.goto(self.A11)
        pen.goto(self.A7)
        pen.end_fill()
        # Ve ngon cay 3/3
        pen.penup()
        pen.goto(self.A10)
        pen.pendown()
        pen.fillcolor('#c2d59b')
        pen.begin_fill()
        pen.goto(self.A12)
        pen.goto(self.A13)
        pen.goto(self.A10)
        pen.end_fill()

        
C0=caythong(x=-300,y=-100,h=313)
C0.paint()
C1=caythong(x=300,y=-100,h=313)
C1.paint()
pen.hideturtle()
turtle.mainloop()
```
